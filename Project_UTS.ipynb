```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca citra grayscale dengan memasukkan path gambar
image = cv2.imread('/content/drive/MyDrive/UTS_PCD_2206152/frame.jpeg', cv2.IMREAD_GRAYSCALE)
# Membaca citra dalam format grayscale

# 2. Hitung histogram
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
# Menghitung histogram citra dengan 256 bin (intensitas pixel 0-255)

# 3. Normalisasi histogram
hist_norm = hist / hist.sum()
# Melakukan normalisasi histogram sehingga jumlah frekuensinya menjadi 1

# 4. Visualisasi gambar dan histogram
plt.figure(figsize=(18, 6))

# Kode program untuk menampilkan hasil Gambar grayscale
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Gambar Grayscale')
plt.axis('off')  # Hilangkan sumbu

# Kode program untuk menampilkan hasil Histogram asli
plt.subplot(1, 3, 2)
plt.plot(hist, color='blue')
plt.title('Histogram Asli')
plt.xlabel('Intensitas Pixel')
plt.ylabel('Frekuensi')

# Kode program untuk menampilkan hasil Histogram ternormalisasi
plt.subplot(1, 3, 3)
plt.plot(hist_norm, color='red')
plt.title('Histogram Ternormalisasi')
plt.xlabel('Intensitas Pixel')
plt.ylabel('Frekuensi Normalisasi')

plt.tight_layout()
plt.show()
```

Dengan melihat histogram, kita bisa menilai kontras dan kecerahan citra, serta mendapatkan gambaran tentang dominasi area tertentu dalam gambar berdasarkan distribusi intensitasnya.

```python
import numpy as np  # Menambahkan pustaka NumPy untuk operasi array
import cv2          # Pustaka OpenCV untuk pengolahan citra
import matplotlib.pyplot as plt  # Pustaka untuk visualisasi gambar

# 1. Membaca gambar berwarna
image_color = cv2.imread('/content/drive/MyDrive/UTS_PCD_2206152/Myprofile.jpeg')
# Membaca citra berwarna menggunakan OpenCV

# 2. Kernel rata-rata 3x3
kernel = np.ones((3, 3), np.float32) / 9
# Membuat kernel rata-rata 3x3 yang akan digunakan untuk konvolusi

# 3. Terapkan konvolusi untuk setiap saluran
channels = cv2.split(image_color)
# Memisahkan citra berwarna menjadi tiga saluran (merah, hijau, biru)
convolved_channels = [cv2.filter2D(ch, -1, kernel) for ch in channels]
# Terapkan konvolusi pada setiap saluran menggunakan filter kernel rata-rata

# 4. Gabungkan kembali saluran setelah konvolusi
convolved_color = cv2.merge(convolved_channels)
# Menggabungkan kembali saluran yang sudah mengalami konvolusi menjadi citra berwarna

# Visualisasi hasil citra asli
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))
# Menampilkan citra asli dengan konversi warna BGR ke RGB untuk tampilan yang benar
plt.title('Citra Asli')  # Judul plot untuk citra asli

# Visualisasi hasil konvolusi berwarna
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(convolved_color, cv2.COLOR_BGR2RGB))
# Menampilkan citra setelah konvolusi dengan konversi warna BGR ke RGB
plt.title('Hasil Konvolusi Berwarna')  # Judul plot untuk citra hasil konvolusi

plt.tight_layout()  # Mengatur tata letak agar rapi
plt.show()  # Menampilkan plot
```

# **Penjelasan:**

Hasil output gambar setelah diberi efek konvolusi dengan kernel rata-rata 3x3 hasilnya yaitu gambarnya menjadi tampak lebih halus (blurred) karena setiap piksel hasil konvolusi dihitung sebagai rata-rata dari 9 piksel di sekitarnya dalam jendela 3x3. Efek ini menyebabkan detail kecil, seperti garis tepi, tekstur halus, atau noise, menjadi berkurang, sehingga gambarnya terlihat lebih lembut. Untuk gambar berwarna, warna pada setiap saluran (R, G, B) akan tampak lebih menyatu, menghasilkan efek warna yang lebih rata. Perubahan ini mengurangi noise dan menghasilkan gambar yang lebih bersih, tetapi dengan ketajamannya berkurang, terutama pada tepi objek atau pola yang rumit, sehingga detail gambar menjadi kurang jelas dibandingkan dengan gambar asli. Jika kernelnya diubah menjadi (5,5) atau lebih tinggi maka hasil gambarnya akan terlihat mejadi lebih tinggi cahayanya, karena  konvolusi menyebarkan intensitas cahaya di area yang lebih luas.

```python
# Import pustaka yang diperlukan
import cv2  # Untuk membaca dan memproses citra
import numpy as np  # Untuk mendukung operasi numerik
import matplotlib.pyplot as plt  # Untuk menampilkan citra

# 1. Baca citra grayscale
image = cv2.imread('/content/drive/MyDrive/UTS_PCD_2206152/night.jpeg', cv2.IMREAD_GRAYSCALE)
# Membaca citra dalam mode grayscale

# 2. Transformasi Fourier
dft = cv2.dft(np.float32(image), flags=cv2.DFT_COMPLEX_OUTPUT)
# Transformasi Fourier diskret untuk mendapatkan spektrum frekuensi
dft_shift = np.fft.fftshift(dft)
# Menggeser komponen frekuensi rendah ke pusat spektrum

# 3. Hitung magnitude spectrum
magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]) + 1)
# Menghitung magnitudo spektrum untuk menampilkan distribusi intensitas frekuensi

# 4. Invers Transformasi Fourier
dft_ishift = np.fft.ifftshift(dft_shift)
# Menggeser kembali komponen frekuensi ke posisi awal
image_reconstructed = cv2.idft(dft_ishift)
# Melakukan invers transformasi Fourier untuk kembali ke domain spasial
image_reconstructed = cv2.magnitude(image_reconstructed[:, :, 0], image_reconstructed[:, :, 1])
# Mengambil nilai magnitudo dari hasil invers transformasi

# 5. Normalisasi hasil invers agar bisa divisualisasikan
image_reconstructed = cv2.normalize(image_reconstructed, None, 0, 255, cv2.NORM_MINMAX)
# Normalisasi untuk memastikan piksel hasil invers berada pada rentang nilai 0–255

# 6. Visualisasi hasil
plt.figure(figsize=(15, 10))  # Mengatur ukuran tampilan plot

# Gambar asli
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
# Menampilkan gambar asli dalam skala abu-abu
plt.title("Citra Asli")  # Judul plot
plt.axis('off')  # Menghilangkan sumbu koordinat

# Magnitude Spectrum
plt.subplot(1, 3, 2)
plt.imshow(magnitude_spectrum, cmap='gray')
# Menampilkan magnitudo spektrum dari transformasi Fourier
plt.title("Magnitude Spectrum")  # Judul plot
plt.axis('off')  # Menghilangkan sumbu koordinat

# Gambar hasil invers
plt.subplot(1, 3, 3)
plt.imshow(image_reconstructed, cmap='gray')
# Menampilkan gambar hasil invers transformasi Fourier
plt.title("Hasil Invers Transformasi Fourier")  # Judul plot
plt.axis('off')  # Menghilangkan sumbu koordinat

plt.tight_layout()  # Mengatur tata letak plot agar lebih rapi
plt.show()  # Menampilkan semua plot
```

# **Penjelasan:**

Hasil gambar dari program menunjukkan tiga visualisasi utama diantaranya yaitu yang pertama, citra asli dalam format grayscale yang merepresentasikan intensitas piksel gambar asli; kedua, magnitude spectrum, yang merupakan hasil Transformasi Fourier, menunjukkan distribusi frekuensi pada citra dengan frekuensi rendah terpusat di tengah dan frekuensi tinggi di pinggir, memberikan informasi tentang detail tekstur dan pola dalam gambar; dan ketiga, gambar hasil invers Transformasi Fourier yang terlihat sangat mirip dengan gambar asli. Hal ini menunjukkan bahwa proses Fourier bersifat reversible, di mana citra dapat dikembalikan ke bentuk semula tanpa kehilangan informasi, selama tidak ada manipulasi signifikan pada data frekuensi.

```python
# Import pustaka yang diperlukan
import cv2  # Untuk membaca dan memproses citra
import numpy as np  # Untuk mendukung operasi numerik
import matplotlib.pyplot as plt  # Untuk menampilkan citra

# Baca citra berwarna
image_color = cv2.imread('/content/drive/MyDrive/UTS_PCD_2206152/gelap.jpeg')  # Membaca citra berwarna dalam format BGR

# Tentukan konstanta untuk pencerahan
brightness_constant = 50  # Nilai konstanta yang ditambahkan untuk meningkatkan kecerahan

# Operasi pencerahan dengan clipping untuk setiap saluran warna
brightened_image_color = cv2.add(image_color, (brightness_constant, brightness_constant, brightness_constant, 0))
# Fungsi cv2.add secara otomatis melakukan clipping nilai piksel agar tidak melebihi 255

# Visualisasi hasil
plt.figure(figsize=(12, 6))  # Mengatur ukuran tampilan plot

# Gambar asli
plt.subplot(1, 2, 1)  # Membuat subplot untuk menampilkan gambar asli
plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))  # Mengubah format BGR ke RGB untuk visualisasi
plt.title("Citra Asli Berwarna")  # Judul untuk citra asli
plt.axis('off')  # Menghilangkan sumbu koordinat

# Gambar setelah pencerahan
plt.subplot(1, 2, 2)  # Membuat subplot untuk menampilkan gambar hasil pencerahan
plt.imshow(cv2.cvtColor(brightened_image_color, cv2.COLOR_BGR2RGB))  # Mengubah format BGR ke RGB untuk visualisasi
plt.title(f"Citra Setelah Pencerahan (+{brightness_constant})")  # Judul untuk citra hasil pencerahan
plt.axis('off')  # Menghilangkan sumbu koordinat

plt.tight_layout()  # Mengatur tata letak agar lebih rapi
plt.show()  # Menampilkan semua plot
```

# **Penjelasan:**

Hasil gambar Setelah pencerahan, intensitas piksel pada citra meningkat secara merata sesuai dengan nilai konstanta yang ditambahkan. Dengan ini membuat citra terlihat lebih terang secara keseluruhan, dengan detail pada area yang sebelumnya gelap menjadi lebih jelas terlihat. Peningkatan intensitas piksel dibatasi oleh operasi clipping, sehingga nilai piksel tidak melewati 255 atau nilai maksimum, yang menyebabkan area yang sudah terang sebelumnya dapat berubah menjadi putih. Perubahan ini meningkatkan visibilitas dan memberikan kesan lebih cerah pada citra, tetapi juga dapat mengurangi kontras di area dengan intensitas tinggi, sehingga detail halus pada area tersebut berpotensi hilang.
